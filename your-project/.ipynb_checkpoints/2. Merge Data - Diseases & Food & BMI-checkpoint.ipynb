```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diseases = pd.read_csv('Data_clean/diseases_final.csv', index_col=0)
diseases.head()
```

```python
food = pd.read_csv('Data_clean/food_final.csv', index_col=0)
food.head()
```

```python
bmi = pd.read_csv('Data_clean/BMI_clean.csv', index_col=0)
bmi.head()
```

```python
diseases.columns
```

```python
diseases = diseases[['Year', 'Diabetes %', 'Malnutrition %', 'Cancer stomach %', 'Cancer oral %', 'Cancer oesophagus %',
                    'Cancer colon %']]
diseases.head()
```

```python
food.columns
```

```python
food = food[['Year', 'All Sugar forms pp', 'Fruits & Veggiess pp', 'Sweetners, Honey, Dates pp']]
food.head()
```

```python
diseases.shape
```

```python
food.shape
```

```python
total = pd.concat([diseases, food[['All Sugar forms pp', 'Fruits & Veggiess pp', 'Sweetners, Honey, Dates pp']]], axis=1)
total.head()
```

```python
total = pd.merge(total, bmi, on='Year')
total.head()
```

```python
total.to_csv('Data_Clean/total_clean.csv')
```
